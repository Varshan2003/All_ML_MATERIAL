```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#With line
x = np.array([-3,-2,-1,0,1,2,3])
y = np.array([9,4,1,0,1,4,9])
plt.plot(x,y)
plt.show()
```

```python
#without line
x = np.array([-3,-2,-1,0,1,2,3])
y = np.array([9,4,1,0,1,4,9])
plt.plot(x,y,'o')
plt.show()
```

```python
#without line
x = np.array([-3,-2,-1,0,1,2,3])
y = np.array([9,4,1,0,1,4,9])
plt.plot(x,y,marker = '*')
plt.show()
```

# SUBPLOT

### The subplot() Function
### The subplot() function takes three arguments that describes the layout of the figure.

### The layout is organized in rows and columns, which are represented by the first and second argument.

### The third argument represents the index of the current plot.

```python
x1 = [1,2,3]
y1 = [1,4,9]
plt.subplot(1,2,1)
plt.plot(x1,y1,marker = '*')

x2 = [1,2,3]
y2 = [1,3,4]
plt.subplot(1,2,2)
plt.plot(x2,y2,marker = '*')

plt.show()
```

```python
x1 = [1,2,3]
y1 = [1,4,9]
plt.subplot(2,1,1)
plt.plot(x1,y1,marker = '*')

x2 = [1,2,3]
y2 = [1,3,4]
plt.subplot(2,1,2)
plt.plot(x2,y2,marker = '*')

plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

# SCATTER PLOT

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#b132')

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

# Bar Graph

```python
critiria = ["Age","SSLC","PUC","Height"]
Varshan = [0.2,0.89,0.93,0.58]
plt.bar(critiria,Varshan)
plt.plot()
```

```python
critiria = ["Age","SSLC","PUC","Height"]
Varshan = [0.2,0.89,0.93,0.58]
plt.bar(critiria,Varshan,color = "black",width = 0.3)
plt.plot()
```

```python
critiria = ["Age","SSLC","PUC","Height"]
Varshan = [0.2,0.89,0.93,0.58]
plt.barh(critiria,Varshan,color = "pink")  #horizontal
plt.plot()
```

# PIE Chart

```python
count = [23,4,52,55,12]
label_new = ["Sebu","Drakshi","ananas","Dalimbe","Sapota"]
plt.pie(count,labels = label_new)
plt.show()
```
